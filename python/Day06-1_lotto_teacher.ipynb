```python
import math

def mean(xs):#평균
    return sum(xs)/len(xs)

def dev(xs): #편차
    m = mean(xs)
    return [x-m for x in xs]

def var (xs): #분산
    n = len(xs)
    d = dev(xs)
    return sum(x**2 for x in d)/(n-1)

def stdev(xs):#표준편차
    return math.sqrt(var(xs))
```

```python
csv_data=[]
with open('lott.CSV') as f:
    for line in f:
#         csv_data.append(line.replace('\n','').split(','))
        csv_data.append(line[:-1].split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

### 문자열 -> 숫자

```python
for game in csv_data: #csv_data속 game 이니까 game은 list type
#     for num in game: # game 속 num이니까 num은 Str
    for i, num in enumerate(game):
        game[i] = int(num)


print(csv_data[:3])
```

```python
# 2차원 배열 도는 방법...in java
for i in range(len(csv_data)):
    for j in range(len(csv_data[i])):
        csv_data[i][j] = int(csv_data[i][j])
        
print(csv_data[:3])
```

#### 2차원 -> 1차원

```python
# flat (2차원을 1차원으로 납작하게('flat'))
lott_num_flat = []
for game in csv_data:
    lott_num_flat = lott_num_flat + game

print(lott_num_flat[:20])
```

```python
lott_num_flat2 = []
for i in range(len(csv_data)):
    for j in range(len(csv_data[i])):
        lott_num_flat2.append(csv_data[i][j])
        
print(lott_num_flat2[:20])
```

##### 전체 회차를 통털어 빈도수(mode, 최빈값): 숫자별 카운트


```python
lott_num_domain = list(set(lott_num_flat)) #중복없애기위해 set으로 바꿔줌 -> 그걸다시 list로 바꿈...
lott_num_freq = []
for num in lott_num_domain:
    x = lott_num_flat.count(num)
    lott_num_freq.append(x)
    
# print(lott_num_domain[:10])
# print(lott_num_freq[:10])

dict_freq = dict(zip(lott_num_domain, lott_num_freq))
dict_freq = dict(sorted(dict_freq.items(), key = lambda x: x[1], reverse = True))

print(dict_freq)

# dict.keys()    #key만 가져옴 => .keys()
x = list(dict_freq.keys())
print(x[:6])
# mode = 43
```

#### 회차별 - 평균값 / 표준편차

```python
for i, game in enumerate(csv_data):
    print('회차:{}, \t평균:{}, \t표준편차:{}'.format(i+1, mean(game), stdev(game)))
```
