```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
pd_data = pd.read_csv(url, header = None)

pd_data
```

```python
pd_data.columns = ['성별','Length','Diameter','Height','전체무게','살점무게','내장무게','껍질무게','Rings']
```

```python
pd_data = pd_data[['성별','전체무게','살점무게', '내장무게', '껍질무게']]
```

```python
pd_data_I = pd_data[pd_data['성별'] == 'I'].index
pd_data.drop(pd_data_I , inplace=True)
```

```python
g1 = pd_data.groupby(['성별'])
df_group = g1.mean()*10
# df_group.reset_index()
df_group_pv = df_group.reset_index().pivot('전체무게','성별')
sns.heatmap(df_group_pv, annot=True)
```

```python
sns.pairplot(pd_data, hue='성별', markers=['o', 's'])
```
