```python
ride = pd_raw.drop(0)
b = ride['계'].astype(np.int64)

ride['계'] = b
ride
```

```python
ride[['계','계.1','계.2','계.3','계.4','관용','관용.1','관용.2','관용.3','관용.4','자가용','자가용.1','자가용.2','자가용.3','자가용.4']] =ride[['계','계.1','계.2','계.3','계.4','관용','관용.1','관용.2','관용.3','관용.4','자가용','자가용.1','자가용.2','자가용.3','자가용.4']].astype(np.int64)
```

```python
ride.dtypes
```

```python
g1 = ride.groupby(['시군구(1)','시점'])
df_group = g1.mean()
df_group.reset_index()
df_group_pv = df_group.reset_index().pivot('시군구(1)','시점', '계')
fig = plt.figure(figsize=(6.4*4, 4.8*4))
sns.heatmap(df_group_pv, annot=True, fmt='.0f')
```

```python
ride['시점'] = pd.to_datetime(ride['시점'], format='%Y-%m')
# 시점에서 년도만 추출하여 새로운 컬럼으로 추가
ride['year'] = ride['시점'].dt.year

# year 컬럼을 추가한 후 groupby하여 각 년도별 평균값 계산
g1 = ride.groupby(['시군구(1)', 'year'])
df_group = g1.mean()

# pivot하여 heatmap으로 시각화
df_group_pv = df_group.reset_index().pivot('시군구(1)', 'year', '계')
fig = plt.figure(figsize=(6.4, 4.8))
sns.heatmap(df_group_pv, annot=True, fmt='.0f')
```

```python
ride['year'] = pd.DatetimeIndex(ride['시점']).year
ride['month'] = pd.DatetimeIndex(ride['시점']).month

g1 = ride.groupby(['시군구(1)', 'year', 'month'])
df_group = g1.mean()

df_group_pv = df_group.reset_index().pivot(index='시군구(1)', columns=['year', 'month'], values='계')
fig = plt.figure(figsize=(6.4*4, 4.8*4))
sns.heatmap(df_group_pv, annot=True, fmt='.0f')
```
