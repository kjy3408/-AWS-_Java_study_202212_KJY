```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)


pd_raw = pd.read_csv('시도별화재발생현황총괄3.csv', encoding='ANSI')

filter1 = pd_raw['행정구역별']!='전국'
pd_raw = pd_raw[filter1]

pd_raw
```

```python
print(type(pd_raw['행정구역별']))
```

```python
fig1 = plt.figure()
a = fig1.add_subplot()
sns.lineplot(data=pd_raw, x='시점', y='행정구역별', hue='사망 (명)')
```

```python
a = pd_raw[['행정구역별','건수 (건)']]
b = a.groupby('행정구역별').sum()
```

```python
plt.xticks(rotation=90)
sns.scatterplot(data=b, x='행정구역별', y='건수 (건)')
```

```python
a = pd_raw[['시점','건수 (건)','사망 (명)']]
b = a.groupby('시점').sum()
b
sns.lineplot(data=b, x='시점', y='사망 (명)')
```

```python
sns.lineplot(data=b, x='시점', y='건수 (건)')
```

```python
# plt.xticks(rotation=270)
sns.lineplot(data=pd_raw, x='시점', y='사망 (명)')
```

```python
a = pd_raw[['행정구역별', '재산피해(계) (천원)', '이재민수 (명)', '동산 (천원)']]
b = a.groupby('행정구역별').sum()

# plt.xticks(rotation=270)


sns.displot(data=b, x='행정구역별', y='이재민수 (명)')
plt.xticks(rotation=45)
```
