```python
list_a = [i for i in range(10)]
list_b = [i for i in range(100, 110)]
print(list_a)
print(list_b)

# in
if 10 not in list_a:
    print('True')
else:
    print('False')

#concat -concatenation (이어붙이다)
list_c = list_a + list_b
print(list_c)

# * - 2번 나오게해라.
print(list_a * 2)      #2 * list_a 
```

```python
list_d = list('abcdefghicjklmn')
print(list_d)
print(list_d[3]) #list_d속 index 3 출력해라.
print(list_d[3:6]) #list_d속 index3부터 6미만까지 출력해라.(n-1)
print(list_d[3:10:2]) #list_d속 한개건너뛰고 2번째부터 출력해라.

print(len(list_d)) #list_d속 데이터 개수 .
print(min(list_d)) #list_d속 제일 아스키코드가 제일 작은것.
print(max(list_d)) #list_d속 제일 아스키코드가 제일 큰것.

c_index = list_d.index('c') #왼쪽부터 c의 index를 출력.
list_d[c_index] = 'x' #어떠안 index에있는 c를 x로 바꿔라.
print(list_d)

c_index1 = list_d.index('c', 3)#c를 찾는데 index 3부터 찾아봐.
print(c_index1)

c_index1 = list_d.index('c', 3, 10)#c를 찾는데 index 3~ 10미만까지 찾아봐
print(c_index1)

print('count', list_d.count('a'))#list_d속에 'a'는 1개있다.

```

```python
s = 'While The Python Language Reference describes the exact syntax and semantics of the Python language, this library reference manual describes the standard library that is distributed with Python. It also describes some of the optional components that are commonly included in Python distributions.'
print(s.lower())

#제각각의 알파벳 빈도수를 알고싶다. a~z까지
#list화 시키면 알파벳으로 다 나눠진다. 해당 list에서 set으로 바꿔도 해당 알파벳의 개수는 남는다.

list_s = list(s.lower()) #소문자로 다 바꾼다.
#print(list_s)
set_s = set(list_s)
#print(set_s)
exclude_list = ['.', ' ', ','] #없애고 싶은 녀석을 exclude list로 만들어놓는다.
for i in set_s:
    if i not in exclude_list:
        print(i, list_s.count(i))
```

```python
list_a = [1,2,3]
list_b = list('abc')
d1 = {'a':1, 'b':2, 'c':3}
print(list_a)
print(list_b)
print(d1)

d2 = dict(zip(list_b, list_a)) #list_a와 list_b를 zip키워드를 사용하여 key, value로 엮어줌.
print(d2)
print(d2.items())#index가 같은것끼리 tuple로 묶어진 상태로 list로 만들어짐.

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

#key list, value list만 있으면 dict로 만들 수 있다.

print(d3['c'])#key값으로 조회
print(d3.get('c')) 
print(d3.get('x'))#없는 key값으로 조회해도 에러가 발생하지않고 None(없는 값)이라고 출력
d3['y'] = 15 #새로운 key , value 생성.
print(d3)
```

```python
#순환(for문 이용하여 전체출력)
for i in d3: #반복 출력하면 key값만 나옴.
    print(i, d3[i]) #이렇게 출력해야 value값도 나옴.
    
print(d3.keys()) #key값만 출력
print(d3.values()) #value 값만 출력

for i in d3.values(): #value 값만 출력.
    print(i)
    
print(d3)

d3['y'] = 20; #'y'key의 value값 변경
print(d3)
```
