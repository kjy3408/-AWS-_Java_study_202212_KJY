```python
csv_data=[]
with open('전국보행자전용도로표준데이터.csv', encoding='cp949') as f:
    for line in f:
        csv_data.append(line.split(','))
        
print(csv_data[:3])
```

```python
for i, e in enumerate(csv_data[0]): #구분 란 index 나옴.
    print(i, e)
```

```python
walk_list = []
for i in csv_data[1:]:
    walk_list.append(i[15])

    
print(walk_list)
```

```python
#lambda
walk_list = [float(x) if x.replace('.','',1).isdigit() else x for x in walk_list]

walk_list = [x / 100 if type(x) == float and x > 100 else x for x in walk_list]

walk_list = [x for x in walk_list if type(x) == float]

mean = sum(walk_list) / len(walk_list)

print(mean)
```

```python
for i in range(len(walk_list)):
    if walk_list[i] == '':
        walk_list[i] = 0

print(walk_list)
```
