```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd_raw = pd.read_csv('전국일방통행도로표준데이터.csv', encoding='cp949')

col_selection = ['시도명', '지정사유', '지정연도', '도로폭', '도로차로수', '보차분리여부']
pd_data = pd_raw[col_selection]

pd_data
```

```python
# print(pd_data.shape)
# print(pd_data.dtypes)

for e in pd_data:
    print(e, '\t',pd_data[e].hasnans) #nan이 있는지 확인. 
```

```python
# 시도명
print(pd_data['시도명'].unique())
print(pd_data['시도명'].hasnans)

# print(pd_raw[pd_raw['시도명']=='10']) #오류치 찾는 필터링.
filter1 = pd_raw['시도명'] =='10'
# pd_data[filter1]['시도명'] = '강원도'  #필터를 사용해서 오류치를 '강원도'로 바꿔줌. (밑과 같은방식 (warning))
pd_data.loc[filter1, '시도명'] = '강원도'  #필터를 사용해서  '시도명'이 '10'인것을 '강원도'로 바꿔줌


#결측치(NaN) : 없음
#오류치 : '10' -> '강원도'
#이상치 : 없음

print(pd_data['시도명'].unique())
print(pd_data['시도명'].hasnans)
```

```python
# 지정사유
print(pd_data['지정사유'].unique())
print(pd_data['지정사유'].hasnans)

#결측치(NaN) -> '불분명'
na_filter = pd_data['지정사유'].isna() #True -> NaN
# print(na_filter.value_counts())
#결측치(NaN) : 있음 -> '사유없음'
pd_data.loc[na_filter, '지정사유'] = '불분명'  #pd_data['지정사유'].filterna('불분명')과 같음.

#오류치: 같은내용 다른이름 정리
def e1(x):
    if '원활' in x:
        return '원활'
    elif '불편' in x:
        return '불편'
    elif '안전' in x:
        return '안전'
    elif '혼잡' in x:
        return '혼잡'
    else:
        return x

pd_data.loc[:, '지정사유'] = pd_data.loc[:, '지정사유'].apply(e1)

#결측치(NaN) -> '불분명'
#오류치: 같은내용 다른이름 정리
# 이상치(outlier): 없음

print(pd_data['지정사유'].unique())
print(pd_data['지정사유'].hasnans)
```

```python
# 지정연도
print(pd_data['지정연도'].unique())
print(pd_data['지정연도'].hasnans)

# 결측치(NaN) --> 0
na_filter = pd_data['지정연도'].isna()  #isna를 na_filter에 넣어줌.
print(na_filter.value_counts())

pd_data.loc[na_filter, '지정연도'] = 0  

# 결측치(NaN) --> 0
#오류치 : 없음 
#이상치 : 없음

print(pd_data['지정연도'].unique())
print(pd_data['지정연도'].hasnans)
```

```python
#도로폭

print(pd_data['도로폭'].unique())
print(pd_data['도로폭'].hasnans)


# print(pd_data['도로폭'].describe()) #describe 메소드사용 이상치 추측 메소드?
q1, q3 = pd_data['도로폭'].quantile([0.25, 0.75])
iqr = q3 - q1
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
print(lower, upper)
filter1 = pd_data['도로폭'] > upper
print(filter1.value_counts())
pd_data.loc[filter1, '도로폭'] = np.NaN

print(pd_data['도로폭'].unique())
print(pd_data['도로폭'].hasnans)

# 결측치(NaN) : 없음
# 오류치 : 없음
# 이상치 : upper 보다 큰 값을 np.NaN으로 교체 후 dropna() 이용하여 이상치 삭제.
```

```python
#도로 차로수
print(pd_data['도로차로수'].unique())
print(pd_data['도로차로수'].hasnans)

#결측치(NaN) : 1개 -> dropna() 해줌
na_filter = pd_data['도로차로수'].isna()
print(na_filter.value_counts())

#오류치 : 60 --> np.NaN 후에 dropna()이용하여 제거
filter1 = pd_data['도로차로수'] == 60
print(filter1.value_counts())

pd_data.loc[filter1, '도로차로수'] = np.NaN

# 이상치: 없음
print(pd_data['도로차로수'].unique())
print(pd_data['도로차로수'].hasnans)

#결측치(NaN) : 1개 -> dropna() 해줌
#오류치 : 60 --> np.NaN 후에 dropna()이용하여 제거
# 이상치: 없음
```
