```python
csv_data = []
with open('전국건강증진센터표준데이터.csv', encoding = 'cp949') as f:
    for line in f:
#         print(line) 
        csv_data.append(line.split(',')) #',' 콤마 단위로 나눠라.
        
print(csv_data[:3]) #index 3개만 출력.
```

```python
name_list = []
for e in csv_data[1:]: #csv_data 순환 [index 0 인건 나오지않음]
#     print(e[2])  # e 의 type은 list(list속의 반복이기 때문) => e의 index 2 출력
#     print(e[2].split()) #e => list, e[2] => 문자열이기때문에 split을 쓸수있음.
#     print(e[2].split()[0]) #e[2].split()에 [0]번 index만 출력 [에러나올수도있음.]
#     print(e[2].split()) #빈값이 있을 수 있으니 확인. [.split()하면 list가 됨]
    if len(e[2]) > 0: #문자열의 길이가 0보다 크다면 으로 조건 걸어주면서 출력
        name_list.append(e[2].split()[0]) #name_list에 e[2].split()의 0번 index에있는 문자열들을 담아줌.
#         print(e[2].split()[0])

print(name_list) 
    
```

```python
for n in name_list:
#     print(n)
    if n[0] == '"': #name_list의 문자열들 중 문자열의 0번 index에 큰따옴표가 들어있다면 1번 index부터 출력해라
        print(n[1:])
    else:
        print(n) #큰따옴표 없으면 걍 출력해라
        
# for n in name_list:        위와같은작업.(lstrip 활용)
#     print(n.lstrip('"'))
```

```python
for i, n in enumerate(name_list): #위와같음
    name_list[i] = n.lstrip('"')
#     print(n.lstrip('"'))
print(name_list[:10])
```

```python
# name_set = set(name_list) #name_list를 set으로 바꿔줌

for n in set(name_list):
    print(n, name_list.count(n))
```

```python
#빈도수 dictionary 반환

#lambda x: x[1]
# def sort_key(x):
#     return x[1]

def unique_count(raw_list, sort = False):
    raw_set = set(raw_list)
    d_si_list = []
    d_num_list = []
    for i in raw_set:
#         print(i, raw_list.count(i))
        d_si_list.append(i)
        d_num_list.append(raw_list.count(i))
        
    di1 = dict(zip(d_si_list, d_num_list))
    if sort == True:
        di1 = dict(sorted(di1.items(), key = lambda x : x[1]))
#         di1 = dict(sorted(di1.items(), key = sort_key))
    return di1
         
unique_count(name_list, True)
```
