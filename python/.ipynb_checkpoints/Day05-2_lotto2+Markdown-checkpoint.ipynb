```python
csv_data=[]
with open('lott.CSV') as f:
    for line in f:
#         csv_data.append(line.replace('\n','').split(','))
        csv_data.append(line[:-1].split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

```python
# 1. 빈도수 (mode): 숫자별 카운트 (몇번이나 나왔나?)
# 2. 회차별 - 평균값, 표준편차
num_list = []
for i in csv_data:
    num_list.append([int(e)for e in i])

result = sum(num_list, [])

num_set = set(result)


#각 회차 숫자별 빈도수 카운트.
klist=[]
vlist=[]
for i in num_set:
#     print(i, result.count(i))
    klist.append(i)
    vlist.append(result.count(i))
    
d1 = dict(zip(klist, vlist))
# print(d1)
d2 = dict(sorted(d1.items(), key = lambda x: x[1], reverse = True))
print(d2)

```

```python
#각 회차별 당첨번호의 평균값 
#1, 12, 13, 17, 27, 34, 43
for i, e in enumerate(num_list): 
    mean_num = int(sum(e)/len(e))
    print(i+1, mean_num)
```

```python
#회차별 표준편차
import math

for i, e in enumerate(num_list):
    mean_num = int(sum(e)/len(e))    
    p = [y-mean_num for y in e]
    n = len(e)
    stdev = math.sqrt(sum([z*z for z in p])/(n - 1))
    print(i+1,stdev)
```
